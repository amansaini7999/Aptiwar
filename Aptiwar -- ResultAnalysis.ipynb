```python
li = []
def Score():
    
    for j in range(0, 101):
        
        Total = 0
        A = 0
        C = 0
        I = 0
        
        key = df[1][j]
        
        for i in range(3, 43):
            if bool(df[i][j] == -1):
                Total = Total+0
            elif (bool(df[i][j] == df1.iloc[i-3][0])):
                Total = Total+2
                A = A+1
                C = C+1
            else:
                Total = Total-1
                A = A+1
                I = I+1                
        li.append([key, Total, A, C, I])
```

```python
Score()
```

```python
li
```

```python
df = pd.DataFrame(li)
```

```python
df = df.rename(columns = {0:'Name', 1:'Total Obtained Marks', 2:'Total Attempted Questions', 3:'Total Corrected Questions', 4:'Total Incorrect Questions'})
df
```
